```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "client name" : ["김땡땡", "박땡땡", "이땡떙", "문땡땡","구땡땡","황땡땡","장땡땡","최땡땡"],
    "product name" : ["짜장면","짜장면","돈까스","비빔밥","김치찌개","비빔밥","쌀국수","치킨"],
    "purchase values" : [3,2,2,5,3,4,3,1]
})

df
```

purchase values 는 평점 느낌으로 이해하자

```python
calc_df = df.groupby(by=['product name'], as_index=False).sum()
calc_df
```

음식 별 평점들의 합

```python
calc_df['purchase values'] /= len(calc_df)
calc_df.sort_values(by='purchase values', ascending=False)
```

평점의 합에서 개수를 나누기

이 그룹에 한해서 비빔밥, 짜장면 순으로 메뉴판을 제작할 수 있고
그룹의 음식 선호도를 알 수 있었다
